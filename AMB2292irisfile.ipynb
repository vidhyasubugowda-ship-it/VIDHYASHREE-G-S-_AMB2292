```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('/iris (1).csv')
print(data)
```

```python
data["Species"].unique()     #dataframe column name unique gives unique value in the column
```

```python
data["Species"].value_counts()    #gives the count of each unique value in the column
```

```python
# separate independent and dependent variable(inpur/output)
x=data.iloc[:,:-1].values
y=data.iloc[:,4].values
print(x)
print(y)
```

```python
#split the dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
#create decision tree classifier oject
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier( criterion ="entropy",max_depth=None,min_samples_split=5,min_samples_leaf=2)#hyperparameter
```

```python
#train the model(fit method)
#model=obj.fit(x_train,y_train)
model.fit(x_train,y_train)
print(model)
```

```python
#testing by using predict method
y_pred=model.predict(x_test)
print(y_pred)
```

```python

from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
print(c)
```

```python
len(y_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#visualization
from sklearn import tree
fn=['sepal_length(cm)','sepal_width(cm)','petal_length(cm)','petal_width(cm)']
cn=['setosa','versicolor','virginica']
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300) #dpi more quality more
tree.plot_tree(model,
              feature_names=fn,  #column name
              class_names=cn,
              filled=True);
```
