```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#create data with missing values   #not a num -nan
data=pd.DataFrame({
    "plot":['a','b','c','d','e'],
    'yield_kg':[2500,np.nan,2800,np.nan,2600],
    'rainfall':[150,200,np.nan,180,170],
    'temperature':[28,30,29,np.nan,27]
})
print("===data with missing values===")
print(data)
#===detecting missing data===
print("\n===detecting missing data===")
print("is null?")
print(data.isnull())
print(f"\n total missing per column:")
print(data.isnull().sum())   #how many nan is there
```

```python
#fill with mean  #fill na is a method to fill information in place of nan
data_mean=data.copy()
print(data_mean["yield_kg"])
data_mean["yield_kg"]=data_mean["yield_kg"].fillna(data_mean["yield_kg"].mean())
print('\n' ,'fill with mean:')
print(data_mean["yield_kg"])
```

```python
# fill with median
data_median=data.copy()
print(data_median["rainfall"])
data_median["rainfall"]=data_median["rainfall"].fillna(data_median["rainfall"].median())
print('\n' ,'fill with median:')
print(data_median["rainfall"])
```

```python
#forward fill(use previous value)
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')#ffill-it replaces missing value,with last valid (non missing)
print("\n forward fill:")                        #value above it in the column
print(data_ffill)
```

```python
#dropping missing data
print("\n===dropping rows===")
data_dropped=data.dropna()
print(data_dropped)
print(f"original rows:{len(data)},after drop:{len(data_dropped)}")
```

```python
import pandas as pd

salary_data=pd.read_csv('/content/Salary_Data (1).csv')
print(salary_data)
```

```python
# slice dataset as descriptive as(independent) and target variable(dependent)
x=salary_data.iloc[:,0:1].values
print(x)
y=salary_data.iloc[:,-1].values
print(y)
```

```python
#splitting dataset into test and train
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
print(salary_data.shape)
print(x_train.shape)   #shape(rows and columns)  test and train 80% and 20% or 70% or 30%
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import numpy as np
x_mean=np.mean(x_train)     #calculate mean value of input and output
print(x_mean)
y_mean=np.mean(y_train)
print(y_mean)
```

```python
# calculate intercept and coefficient
n=len(x_train)
num,den=0,0
for i in range(n):
  num+=(x_train[i]-x_mean)*(y_train[i]-y_mean)
  den+=(x_train[i]-x_mean)**2

b1=num/den
b0=y_mean-(b1*x_mean)
print(" b0:",b0)
```

```python
# predicting results on test dataset
y_pred_code=b0+b1*x_test
print(y_pred_code)
```

```python
print(y_test)
```

```python
# mean square error
n1=len(y_test)
mse=0
for i in range(n1):
  mse+=(y_test[i]-y_pred_code[i])**2
mse=mse/n1
print(mse)
# root mean square error
rmse=np.sqrt(mse)
print(rmse)
```

```python
#create a object of the algiorithum/model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
#train the model
reg.fit(x_train,y_train)
```

```python
# predict the model on testing daaset
y_pred=reg.predict(x_test)
print(y_pred)
```

```python
# calcu;ate intercept and coefficient
print("intercept",reg.intercept_)
print("coefficient",reg.coef_)
```

```python
# evalution metrics
from sklearn import metrics
print("mean absolute error:")
print(metrics.mean_absolute_error(y_test,y_pred))
print("mean squared error:")
print(metrics.mean_squared_error(y_test,y_pred))
print("root mean squared error:")
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#visualization
import matplotlib.pyplot as plt

plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,reg.predict(x_train),color='blue')
plt.title("salary vs experience(training set)")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```
