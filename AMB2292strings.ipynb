```python
import random
inventory={}
total_money=0

fish_prices={"rohu":12,"catla":18,"sinchu":25}
def showmenu():
     print("1 catch fish")
     print("2 view inventory")
     print("3 sell fish")
     print("4 exit")

def catch_fish():
    fish=random.choice(list(fish_prices.keys()))
    inventory[fish]=inventory.get(fish,0)+1
    print("you caught a {fish}!")
def show_inventory():
    if inventory=={}:
     print("no fish")
    else:
     print("your inventory:")
     for fish,count in inventory.items():
       print(f"{fish}:{count}")
def sell_fish():
  global total_money
  if inventory=={}:
      print("no fish")
  else:
    earnings=0
    for fish,count in inventory.items():
       earnings+=fish_prices[fish]*count
    total_money+=earnings
    inventory.clear()
    print(f"you sold {count} fish for {earnings}")
while True:
  choice=int(input("enter your choice"))
  if choice==1:
      catch_fish()

  elif choice==2:
      show_inventory()
  elif choice==3:
      sell_fish()
  elif choice==4:
      print("total money earned:",total_money)
      print("thanks for playing")
      break
```
