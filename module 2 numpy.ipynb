```python
import numpy as np
 #crop yield for 4 farms over 5 seasons(in tons)
crop_yields=np.array([
    [12.5, 13.2,14.1,13.8, 15.2],
    [10.3, 11.5,10.8,12.1,11.9],
    [15.1,14.8,16.2,15.5,16.8],
    [11.2,10.9,12.3, 11.7,12.8]
])
print("crop yields array (tons):")
print(crop_yields)
print(f'/n shape:{crop_yields.shape}')
print(f'/n data type:{crop_yields.dtype}')
print(f'/n total elements:{crop_yields.size}')
print(f'/n num of dimensions:{crop_yields.ndim}')
```

```python
# agricultural analysis operation
total_yield=np.sum(crop_yields)
avg_yield_per_farm=np.mean(crop_yields,axis=1)#along the rows
avg_yield_per_season=np.mean(crop_yields,axis=0) #along the columns
avg_yield_per_farm=np.median(crop_yields,axis=1)
avg_yield_per_season=np.median(crop_yields,axis=0)
avg_yield_per_farm=np.std(crop_yields,axis=1)
avg_yield_per_season=np.std(crop_yields,axis=0)
max_yield=np.max(crop_yields)
min_yield=np.min(crop_yields)

print(f'/n total production:{total_yield} tons')
print(f'avg per farm:{avg_yield_per_farm} tons')
print(f'avg per season:{avg_yield_per_season} tons')
print(f'max yield:{max_yield} tons')
print(f'min yield:{min_yield} tons')
```

```python
best_farm=np.argmax(avg_yield_per_farm)  #index of maximum value
print(f'/n best performing farm:farm',best_farm+1) # index start from 0 thats y +1 added
print("their avg yield is:", round (avg_yield_per_farm[best_farm],2),"tons")

print(f'/n best performing farm:farm {best_farm+1}')
print(f'/n their avg yield is:{avg_yield_per_farm[best_farm]:.2f} tons')


```

```python
median_yield=np.median(crop_yields)
print(f'/n median yield:{median_yield} tons')
std_yield=np.std(crop_yields)
print(f'/n standard deviation of yield:{std_yield} tons')
```
