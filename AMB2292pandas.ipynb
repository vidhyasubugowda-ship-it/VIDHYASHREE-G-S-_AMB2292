```python
import pandas as pd
data={
    "farm":["farmA","farmB","farmC","farmD"],
    "crop":["wheat","rice","wheat","maize"],
    "area_hectares":[10,15,8,12],
    "production":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.head()     #top rows
```

```python
df.head(2)
```

```python
df.tail(2)  #bottom rows
```

```python
df.info()  #gives basic information about prg
```

```python
df.describe()  #basic summary in the form of stastics
```

```python
df["yield_ton_per_hectare"]=df["production"]/df["area_hectares"]   #adding a new column to the existing one=
print(df)
```

```python
df.iloc[[1,3]]
```

```python
df.iloc[[1,3],[0,1,2]]     #itwrks with only integers not string
```

```python
df.loc[[1,3],["farm","crop","area_hectares"]]    # mixing is done for  both integer and string
```

```python
print(df["crop"]=="wheat")
print()                              #i want only wheat rows and storte it as wheat_farm
wheat_farm=df[df["crop"]=="wheat"]
print(wheat_farm)
```

```python
#farms with yield greater than 3
high_yield=df[df["yield_ton_per_hectare"]>3]
print(high_yield)
```

```python
#sort the forms by highest production
sorted_df=df.sort_values(by="production",ascending=False)
print(sorted_df)
```

```python
#groupby operation -average yield per crop
avg_yield=df.groupby("crop")["yield_ton_per_hectare"].mean()
print(avg_yield)
avg_yield=df.groupby("crop")["yield_ton_per_hectare"].sum()
print(avg_yield)
```
