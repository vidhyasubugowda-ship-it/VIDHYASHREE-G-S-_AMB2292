```python
n = 5 # You can change this value to test different numbers
fact=1
if(n>0):
  for i in range(1,n+1):
   fact=fact*i
  print("factorial of n is",fact)
elif n == 0:
  print("factorial of 0 is 1")
else:
  print("factorial cannot be calculated for negative numbers")
```

```python
def fact(m):
    fact=1
    if (m>0):
        for i in range(1,m+1):
            fact=fact*i
        print("factorial of m is",fact)          #with parameter without return type
    elif m == 0:
     print("factorial of 0 is 1")
n=int(input("enter a number"))
fact(n)
```

```python
#with parameter with return type
def fact(m):
    fact=1
    if (m>0):
        for i in range(1,m+1):
            fact=fact*i
        return fact      # return instead of print
    elif m == 0:
     print("factorial of 0 is 1")
n=int(input("enter a number"))
z=fact(n)#store the returned value in z
print(z)
```
