```python
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print("\nyearly production:")
print(yearly_production)
```

```python
#Scatter plot _rainfall v/s production
import matplotlib.pyplot as plt
plt.scatter(df["Rainfall_mm"],df["Production_Tons"])
plt.xlabel("Rainfall(mm)")
plt.ylabel("Production(Tons)")
plt.title("Rainfall v/s Production")
plt.show()
```

```python
# create bar graph
plt.figure(figsize=(8,5))
plt.bar(yearly_production.index,yearly_production.values,color="brown",width=0.3)
plt.xlabel("Year")
plt.ylabel("Total Production(Tons)")
plt.title(" total crop Production per year")
plt.xticks(yearly_production.index)
plt.show()
```

```python
# bar chart -area by year
area_by_year=df.groupby("Year")["Area_Hectares"].sum()
plt.figure(figsize=(5,5))
plt.bar(area_by_year.index,area_by_year.values,color="black",width=0.3)
plt.xlabel("Year")
plt.ylabel("Total Area(Hectares)")
plt.title("Total Area by Year")
plt.show()
```

```python
wheat=df[df["Crop"]=="Wheat"]
rice=df[df["Crop"]=="Rice"]
plt.figure()
#plot separately
plt.plot(wheat["Year"],wheat["Production_Tons"],label="Wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="Rice")
plt.xlabel("Year")
plt.ylabel("Production(Tons)")
plt.title("crop production over years")
plt.legend()
plt.show()
```

```python
data_2022=df[df["Year"]==2022]
print(data_2022)
plt.figure(figsize=(6,4))
plt.bar(data_2022["Crop"],data_2022["Production_Tons"],color=["green","orange"])
plt.xlabel("Crop")
plt.ylabel("Production(Tons)")
plt.title("Crop Production in 2022")
plt.legend()
plt.show()
```

```python
#Customstyled bar plot
plt.figure(figsize=(6,4))
bars=plt.bar(yearly_production.index,yearly_production.values,color="blue",edgecolor="black")
#add value label on top of bars
for bar in bars:
  height=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,
           height,
           str(height),
           ha="center",
           va="bottom")

  plt.xlabel("Year")
  plt.ylabel("Total Production(Tons)")
  plt.title("yearly_production with values")
plt.show()
```
