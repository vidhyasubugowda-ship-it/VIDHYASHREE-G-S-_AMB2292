# 1: Data Structure - List

### 1.1: List declaration

```python
# Method 1

L = ["Pratham", 'Sharma', 3.14, 3]

print ("L                  -->", L)
print ("Number of elements -->", len(L))
print ("Type of Object     -->", type(L))
```

```python
# Method 2

L = list(["Pratham", 'Sharma', 3.14, 3])
#L = list(("Pratham", 'Sharma', 3.14, 3)) # Also works 

print ("L                  -->", L)
print ("Number of elements -->", len(L))
print ("Type of Object     -->", type(L))
```

### 1.2: List iteration using while loop

```python

L = ["Pratham", 'Sharma', 3.14, 3]
print ("L -->", L)

print ("")
i = 0
while i < len(L): 
    print (L[i])
    i += 1
```

### 1.3: List iteration using for loop

```python
# Example 1:

L = ["Pratham", 'Sharma', 3.14, 3]
print ("L -->", L)

print ("")
for i in range(0, len(L)):
    print (L[i])
```

```python
# Example 2:

L = ["Pratham", 'Sharma', 3.14, 3]
print ("L -->", L)

print ("")
for s in L:
    print (s)
```

### 1.4: Accessing/Selecting List item

```python
# Example 1:

L = [3, 6, 9, 12, 5, 3, 2]
print ("L     -->", L)

print ("")
print ("L[1]  -->", L[1])
print ("L[2]  -->", L[2])
print ("L[-1] -->", L[-1])
print ("L[-2] -->", L[-2])
```

```python
# Example 2:

L = [3, 6, 9, 12, 5, 3, 2]
print ("L        -->", L)

print ("")
print ("L[1:3]   -->",L[1:3])
print ("L[2:]    -->",L[2:])
print ("L[2:5]   -->",L[2:5])
print ("L[:2]    -->",L[:2])
print ("L[:-1]   -->",L[:-1])
print ("L[-4:-1] -->",L[-4:-1])
```

```python
# Example 3:

L = [3, 6, 9, 12, 5, 3, 2]
print ("L        -->", L)

print ("")
print ("L[::1]   -->", L[::1])        # No change
print ("L[::-1]  -->", L[::-1])       # Reverse
print ("L[::-2]  -->", L[::-2])
print ("L[::2]   -->", L[::2])
```

### 1.5: Add element to list using append()

```python

L = ["Pratham", 'Sharma', 3.14, 3]
print ("L -->", L)

L.append("Rahul")                 # Add element at last
print ('\nAfter Adding "Rahul" -->', L)

L.append("Anmol")                 # Add element at last
print ('\nAfter Adding "Anmol" -->', L)
```

### 1.6: Inserting element in list using insert()

```python

L = ["Pratham", 'Sharma', 3.14, 3]
print ("L -->", L)

L.insert(2, "Rahul")                        # Insert element at 2
print ("\nAfter Insertion 'Rahul'   -->", L)

L.insert(len(L), "Thapar")                  # Insert element at last
print ("After Insertion 'Thapar'  -->", L)           

L.insert(-1, "SIH")                         # Insert element at -1  
print ("After Insertion 'SIH'     -->", L)           

L.insert(10, "Sanchit")                     # No error; Element will be added at last
print ("After Insertion 'Sanchit' -->", L)
```

### 1.7: Add multiple elements to list using extend()

```python
# Example 1

L = ["Pratham", 'Sharma', 3.14, 3]
print ("L -->", L)

L.extend(["Rahul"])                        # Add element at last
print ("\nAfter Adding 'Rahul' -->", L)
```

```python
# Example 2

L = ["Pratham", 'Sharma', 3.14, 3]
print ("L -->", L)

L.extend(["Rahul", "Thapar", 100, 200])    # Add element at last
print ("\nAfter Adding L -->", L)
```

### 1.8: Deleting item from list using del

```python

L = ["Pratham", 'Sharma', 3.14, 3, 6, 7, 8]
print ("L                   -->", L)

del L[1]                                # Delete element at 1
print ("Delete element at 1 -->", L)

del L[1]                                # Delete element at 1
print ("Delete element at 1 -->", L)

del L[-1]                               # Delete last element
print ("Delete element at -1-->", L)

del L[15]                               # Error; index out of range; Exception should me handle
print ("Delete element at 15 -->", L)


# Multiple elements cannot be remove using del
```

### 1.9: Deleting from list using remove()

```python

L = ["Pratham", 'Sharma', 3.14, 3, 6, 7, 8]
print ("L                -->", L)

L.remove(3)                             # Remove element 3
print ("Remove element 3 -->", L)

L.remove(6)                             # Remove element 6
print ("Remove element 6 -->", L)

L.remove(9)                             # Remove element 9 
print ("Delete element at 9 -->", L)    # Error; 9 is not present in L

# Multiple elements cannot be remove using remove
```

### 1.10: Deleting from list using pop()

```python

# Pop: Delete last element from list

L = ['A', 'B', 'C', 'D']
print ("List           -->", L)

L.pop()                         # Delete last element
print ("1. After pop   -->", L)

L.pop(1)                        # Delete element of location 1
print ("2. After pop(1)-->", L)

L.pop()                         # Delete last element
print ("3. After pop   -->", L)

L.pop()                         # Delete last element
print ("4. After pop   -->", L)

L.pop()                         # Error; List is empty; Underflow
print ("5. After pop   -->", L)

#Note: If list is empty, pop() will raise an exception
```

### 1.11: Deleting list

# 2: Data Structure - Dictionary
##### Key-value pair

### 2.1: Dictionary declaration

```python

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}

print ("Dictionary     -->", CGPA)
print ("Total Elements -->", len(CGPA))
print ("Type of Object -->", type(CGPA))
```

### 2.2: Access Dictionary item

```python

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}

print ("Dictionary     -->", CGPA)
print ("Total Elements -->", len(CGPA))

print ("\nCGPA of 1 -->", CGPA[1])  # Pass the Key and you will get value
print ("CGPA of 4 -->", CGPA[4])
print ("CGPA of 7 -->", CGPA[7])
print ("CGPA of 3 -->", CGPA[3])  # Error; Key is not available; Exception should me handle
```

### 2.3: Traverse dictionary

```python
CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}

for k in CGPA:
    print ("CGPA of ", k, " -->", CGPA[k])
```

### 2.4: Getting Keys and Values

```python
CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary -->", CGPA)

print ("Keys       -->", CGPA.keys())
print ("Values     -->", CGPA.values())
```

### 2.5: Adding element in Dictionary

```python

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary        -->", CGPA)

CGPA[3] = 8.6            # Adding
print ("\nAfter Adding(3)   -->", CGPA)

CGPA[9] = 7.6            # Adding
print ("After Adding(9)   -->", CGPA)
# Order or Sequence don't matter
```

### 2.6: Updating element in Dictionary

```python

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary        -->", CGPA)

CGPA[4] = 9.2            # Updating
print ("\nAfter Updating(4) -->", CGPA)

CGPA[10] = 8.8            # Adding
print ("After Adding(10)  -->", CGPA)
```

### 2.7: Deleting element from Dictionary

```python

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary         -->", CGPA)

del CGPA[1]              # Deleting Key 1
print ("\nAfter Deleting (1) -->", CGPA)

del CGPA[8]              # Deleting Key 8
print ("After Deleting (8) -->", CGPA)

del CGPA
print ("After Deleting -->", CGPA)
```

### 2.8: Clear elements in Dictionary

```python

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary             -->", CGPA)

CGPA.clear()            # Deleting all elements
print ("\nDictionary after clear -->", CGPA)
```

### 2.10: Sum/Average all values of Dictionary

```python
# Example 1:
CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary -->", CGPA)

s = 0
for i in CGPA:  
    s = s + CGPA[i]

print ("\nSum        -->", s)
print ("Average    -->", s/len(CGPA))
print ("Average    -->", s//len(CGPA))
```

```python
# Example 2:

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary --> ", CGPA)
#del sum
print ("\nSum        -->", sum(CGPA.values()))
print ("Average    -->", sum(CGPA.values())/len(CGPA))
print ("Average    -->", sum(CGPA.values())//len(CGPA))
```

### 2.11: Min/Max values in the Dictionary

```python

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary -->", CGPA)

print ("\nMin value  -->", min(CGPA.values()))
print ("Max value  -->", max(CGPA.values()))
```

### 2.12: Find key having min/max value in the Dictionary

```python
CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary --> ", CGPA)

print ("\nMin value  -->", min(CGPA.values()))
print ("Max value  -->", max(CGPA.values()))

# method 1 to find Key for Min value

keyList = list(CGPA.keys()) 
valList = list(CGPA.values()) 

minval=min(CGPA.values())
print ("\nValue =", minval," Key -->", keyList[valList.index(minval)])

# method 2 to find Key for Min value
print ("\nKey having Min value  -->", min(CGPA, key = CGPA.get))
print ("Key having Max value  -->", max(CGPA, key = CGPA.get))
```

### 2.13: Find key for any value in the Dictionary

```python

CGPA = {1:8.9, 2:5.6, 4:6.7, 7:9.1, 8:5.3}
print ("Dictionary       -->", CGPA)

# List out keys and values separately 
keyList = list(CGPA.keys()) 
valList = list(CGPA.values()) 

print ("\nValue = 5.6; Key -->", keyList[valList.index(5.6)])
print ("Value = 9.1; Key -->", keyList[valList.index(9.1)])


# one-liner 
print ("\nValue = 5.6; Key -->",list(CGPA.keys())[list(CGPA.values()).index(5.6)])
print ("Value = 9.1; Key -->",list(CGPA.keys())[list(CGPA.values()).index(9.1)])
```

### 2.16: Merge Dictionary

```python
# Example 1 - Merge dictionaries

CGPA = {1:8.9, 2:5.6, 4:9.7}
CGPA1= {4:6.7, 7:9.1, 8:5.3}
print ("CGPA               -->", CGPA)
print ("CGPA1              -->", CGPA1)

CGPA1.update(CGPA)

print ("\nCGPA1 after update -->", CGPA1)
print ("\nCGPA after update -->", CGPA)
```
